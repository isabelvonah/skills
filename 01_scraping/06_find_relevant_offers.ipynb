```python
import pandas as pd

data = pd.read_csv("05_offers_rake.csv")
```

```python
import json

with open('thesaurus.json') as json_file:
    thesaurus = json.load(json_file)
```

```python
data.info()
```

# Text-Ansatz

```python
skill_found_text = []

for i in range(len(data)):
    
    counter = 0
    skills =[]
    
    for key in thesaurus.keys():
        if key in data["text"].iloc[i]:
            counter += 1
            skills.append(key)
        for j in range(len(thesaurus[key])):
            if " " + thesaurus[key][j] + " " in data["text"].iloc[i]:
                counter += 1
                skills.append(thesaurus[key][j])
        
    skill_found_text.append([counter, skills])
```

```python
skill_found_text
```

```python
numb = []
for skill in skill_found_text:
    numb.append(skill[0])
    
skills = []
for skill in skill_found_text:
    skills.append(skill[1])
```

```python
data["number_of_skills_text"] = numb
data["skills_found_text"] = skills
```

```python
comparison = data[["text", "number_of_skills_text", "skills_found_text"]]
comparison
```

```python
comparison[comparison["number_of_skills_text"]  > 0]
```

# Keyphrases-Ansatz

```python
data["keyphrases"].iloc[1000]
```

```python
skill_found_keyphr = []

for i in range(len(data)):
    
    counter = 0
    skills =[]
    
    for key in thesaurus.keys():
        for j in range(len(thesaurus[key])):
            if "'" + thesaurus[key][j] + "'" in data["keyphrases"].iloc[i]:
                counter += 1
                skills.append(thesaurus[key][j])
        
    skill_found_keyphr.append([counter, skills])
```

```python
numb = []
for skill in skill_found_keyphr:
    numb.append(skill[0])
    
skills = []
for skill in skill_found_keyphr:
    skills.append(skill[1])
```

```python
data["number_of_skills_keyphr"] = numb
data["skills_found_keyphr"] = skills
```

```python
comparison2 = data[["text", "keyphrases", "number_of_skills_text", "skills_found_text", "number_of_skills_keyphr", "skills_found_keyphr"]]
comparison2
```

```python
comparison2[comparison2["number_of_skills_keyphr"]  > 0]
```

# Kombination

```python
combi = comparison2[(comparison2["number_of_skills_keyphr"]  > 0 )| (comparison2["number_of_skills_text"] > 0)]
combi
```

```python
combi[(combi["number_of_skills_text"] != 0) & (combi["number_of_skills_keyphr"] != 0)]
```

```python
combi[combi["skills_found_text"] == combi["skills_found_keyphr"]]
```

```python
def to_1D(series):
    return pd.Series([x for _list in series for x in _list])
```
