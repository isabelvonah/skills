```python
files = os.path.join("../01_scraping/01_scrape_meta", "result*.csv")
files = glob.glob(files)

numbers = []

for file in files:
    content = pd.read_csv(file)
    numbers.append([content.iloc[0,1], content.iloc[0,2], len(content)])

quantities = pd.DataFrame(numbers, columns = ["cat", "subcat", "number"])
quantities.sort_values("number")
```

```python
conc = pd.concat(map(pd.read_csv, files), ignore_index=True)
print(len(conc))
```

```python
conc_ohne_dupl = conc.drop_duplicates(subset="id")
print(len(conc_ohne_dupl))
```

```python
duplicates = conc.value_counts("id")
duplicates = conc['id'].value_counts()[conc['id'].value_counts()> 1]
duplicates
```

```python
combinations = []

for duplicate in duplicates.index:
    combinations.append(conc[conc["id"] == duplicate].sort_values("subcat")["subcat"].tolist())
```

```python
combinations
```
