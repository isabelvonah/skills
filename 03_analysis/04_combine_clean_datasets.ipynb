```python
import pandas as pd
```

# Ausgangslage
Die beiden importierten csv-Dateien enthalten alle gescrapeten Daten und müssen im folgenden noch zusammengeführt werden.

```python
metadata = pd.read_csv("../01_scraping/02_eliminate_duplicates.csv")
offers = pd.read_csv("../01_scraping/03_scrape_offers.csv")
```

Der erste Versuch ist ein Mergen über das Attribut "id". Ein gutes Zeichen ist, dass die Anzahl der Einträge mit derjenigen von "offers" übereinstimmt. Somit wurden zu allen Angeboten passende Metadaten gefunden.  

```python
merged = metadata.merge(offers, on="id")
merged.info()
```

```python
merged.head(5)
```

## Weglassen von Attributen
### Suchparameter
Als erstes können einige Attribute weggelassen werden, da sie aufgrund des Merging keine Aussage mehr treffen. Kam ein Jobangebot zum Beispiel in mehreren Kategorien vor, erscheint hier nur die erste Erwähnung und die Suchparameter "cat" und "subcat" spielen somit keine Rolle mehr, ebenso wie "page". "employment-typ" und "industry" waren bei allen Duplikaten gleich und sind somit weiterhin aussagekräftig.
### promo
Dieses Attribut wurde gesetzt, wenn die Jobausschreibung speziell mit dem Label gekennzeichnet war. Da diese aber schon auf die jeweilige Suche zugeschnitten sind und durch das Weglassen von ID-Duplikate vielleicht auch bereits fehlerhafte Einträge vorhanden sind, entscheiden wir uns, diese Information nicht weiter zu berücksichtigen.
### date_x, today
Diese Attribute dienten zum Loggen des Scraping-Prozesses und ist nach der Eliminierung der Duplikate nicht mehr relevant.

```python
merged.drop(labels=["cat", "subcat", "page", "promo", "date_x", "today"], axis="columns", inplace=True)
merged
```

## erste Schwierigkeit
Beim Resultat fallen gewisse Zeilen speziell auf. Sie enthalten NaN-Werte in der Spalte "place_x" und bei "company_x" ist bei jenen Offers fälschlicherweise der Ort hinterlegt. Diese Grupp muss genauer untersucht werden:

```python
merged[merged["place_x"].isna()]
```

Es zeigt sich, dass die Werte aus dem zweiten Datensatz insgesamt wohl verlässlicher sind. Bei den ersten beiden Einträgen ist jedoch der Ort unvollständig und der Name des Unternehmens gar nicht vorhanden. Es stellt sich die Frage, ob es viele solche Fälle gibt.

```python
error_rows = merged["place_x"].isna() & merged["company_y"].isna()
merged.loc[error_rows]
```

Es handelt sich um 150 Einträge. Da die restlichen Daten (v.a. Titel / Text) okey sind, entscheiden wir uns, sie trotzdem zu behalten. Ausserdem wäre es wichtig, wenn schon der Company-Name verloren ist, "place_y" mit "company_x" zu ersetzen.

```python
merged.loc[error_rows, ["company_x","place_y"]]
```

```python
merged.loc[error_rows, ["place_y"]] = merged.apply(lambda x: x['company_x'], axis=1)
merged.loc[error_rows, ["company_x","place_y"]]
```

Vollständigkeitshalber werden auch die fehlerhaften Werte in company_x auf NaN gesetzt:

```python
import numpy as np
merged.loc[error_rows, ["company_x"]] = np.NaN
merged.loc[error_rows, ["company_x","place_y"]]
```

Nun wird überprüft, ob in der Spalte "place_y" nur valide Werte vorhanden sind.

```python
merged[merged["place_y"].isna()]
```

```python
merged["place_y"]
```

Somit kann "place_x" entfernt werden.

```python
merged.drop(labels=["place_x"], axis="columns", inplace=True)
merged.rename(columns = {"place_y": "place"}, inplace=True)
merged
```

Nun soll bei allen weiteren überschneidenden Attributen die Übereinstimmung überprüft werden.

```python
merged[merged["title_x"] != merged["title_y"]]
```

Hier scheint es sich nur um Formulierungsdetails zu handeln und deshalb wird nur "title_y" übernommen.

```python
merged.drop(labels=["title_x"], axis="columns", inplace=True)
merged.rename(columns = {"title_y": "title"}, inplace=True)
merged
```

```python
merged[merged["company_x"] != merged["company_y"]]
```

Diejenigen Einträge die zweimal NaN aufweisen sind die zuvor bearbeiteten.

```python
merged[merged["company_x"].isna()]
```

Bei den weiteren scheint ebefalls ein Fehler bei "company_x" unterlaufen zu sein, "company_y" scheinen jedoch valide Werte zu liefern, denn es handelt sich ebenfalls um diese 150 Einträge, bei denen der company-Name nicht mehr verfügbar ist.

```python
merged[merged["company_y"].isna()]
```

Somit kann hier ebenfalls "company_y" übernommen werden.

```python
merged.drop(labels=["company_x"], axis="columns", inplace=True)
merged.rename(columns = {"company_y": "company"}, inplace=True)
merged
```

Mit der Änderung des letzten überbliebenen "date_y" ist der Datensatz somit auf die verbleibenden und relevanten Spalten reduziert.

```python
merged.rename(columns = {"date_y": "date"}, inplace=True)
merged
```
